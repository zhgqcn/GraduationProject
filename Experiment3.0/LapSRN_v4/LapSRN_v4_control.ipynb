```
import os
from PIL import Image

os.chdir("/content/drive/My Drive/LapSRN/LapSRN-r") 
!python test.py \
--input "/content/drive/My Drive/TestImg/val/95-512pix-speed7-ave1.tif" \
--model '/content/drive/My Drive/TestImg/H_Adagrad_epoch_g_500.pth' \
--outputHR2 "/content/drive/My Drive/TestImg/OriginalDataSet_64/output95_X2_R.tif" \
--outputHR4 "/content/drive/My Drive/TestImg/OriginalDataSet_64/output95_X4_R.tif" \
--cuda  

os.chdir("/content/drive/My Drive/LapSRN/LapSRN-g") 
!python test.py \
--input "/content/drive/My Drive/TestImg/val/95-512pix-speed7-ave1.tif" \
--model '/content/drive/My Drive/TestImg/H_Adagrad_epoch_g_500.pth' \
--outputHR2 "/content/drive/My Drive/TestImg/OriginalDataSet_64/output95_X2_G.tif" \
--outputHR4 "/content/drive/My Drive/TestImg/OriginalDataSet_64/output95_X4_G.tif" \
--cuda

out_img_b = Image.open('/content/drive/My Drive/TestImg/all_black_2048.tif') # 全黑色通道
out_img_r = Image.open('/content/drive/My Drive/TestImg/OriginalDataSet_64/output95_X4_R.tif')
out_img_g = Image.open('/content/drive/My Drive/TestImg/OriginalDataSet_64/output95_X4_G.tif')
out_img = Image.merge('RGB', [out_img_r, out_img_g, out_img_b]).convert('RGB')
out_img.save('/content/drive/My Drive/TestImg/result/X4/H_Adagrad_X4_sfdsfsfjsdfjsdlfjkls.tif')
print("saved")
```

## 通道合并

```
# 指定当前的工作文件夹
import os
# 此处为google drive中的文件路径,drive为之前指定的工作根目录，要加上
os.chdir("/content/drive/My Drive/LapSRN_v3/LapSRN_v3.1/") 

from PIL import Image                                 # 注意修改文件夹以及图像名
out_img_r = Image.open('./LapSRN-r/95_X4_Epoch400_R.tif')  # r通道
out_img_g = Image.open('./LapSRN-g/95_X4_Epoch400_G.tif')  # g通道
out_img_b = Image.open('/content/drive/My Drive/TestImg/all_black_2048.tif')      # b通道
out_img = Image.merge('RGB', [out_img_r, out_img_g, out_img_b]).convert('RGB')    # 合并
out_img.save('./H_95_Epoch400_X4.tif')     # 保存
print('saved')
```
